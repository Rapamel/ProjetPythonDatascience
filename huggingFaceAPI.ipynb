```python
def nthparquet(n):
    request=requests.get(url+f"/parquet/default/train/{n}.parquet",headers=headers)
    if request.ok:
        buffer=BytesIO(request.content)
        table=pq.read_table(buffer)
        return table.to_pandas()
df0=nthparquet(0)
```
